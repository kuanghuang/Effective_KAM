```python
import sympy as sp
from sympy import *
```

```python
x1 = sp.symbols('x1')
x2 = sp.symbols('x2')
y1 = sp.symbols('y1')
y2 = sp.symbols('y2')
eps = sp.symbols('epsilon')

H = (y1**2 + y2**2) / 2 + eps**2/2 * (y1*sin(x1) + y2*sin(x2))**2 / (1 + eps**2 * (sin(x1)**2 + sin(x2)**2))
```

```python
H
```

```python
X_H = Matrix([diff(H, y1), diff(H, y2), -diff(H, x1), -diff(H, x2)])
X_H
```

```python
z1 = sp.symbols('z1')
z2 = sp.symbols('z2')
z3 = sp.symbols('z3')
z4 = sp.symbols('z4')
phi = sp.symbols('phi')

F = X_H.subs(x1, x1+z1)
F = F.subs(x2, x2+z2)
F = F.subs(y1, 1+z3)
F = F.subs(y2, phi+z4)
F = simplify(F)
F
```

```python
point = {z1: 0, z2: 0, z3: 0, z4: 0}
z_vars = [z1, z2, z3, z4]

# Zeroth order: F(0)
F0 = F.subs(point)

# First order: linear terms in z's
F1 = Matrix.zeros(4, 1)
for i in range(4):
    for zi in z_vars:
        F1[i] += F[i].diff(zi).subs(point) * zi

# Second order: quadratic terms (diagonal + cross terms)
F2 = Matrix.zeros(4, 1)
for i in range(4):
    # Diagonal terms:  Fi/zj * zj
    for j, zj in enumerate(z_vars):
        F2[i] += 0.5 * F[i].diff(zj, 2).subs(point) * zj**2
    # Cross terms: Fi/zjzk * zj*zk for j < k
    for j, zj in enumerate(z_vars):
        for k in range(j+1, 4):
            zk = z_vars[k]
            F2[i] += F[i].diff(zj, zk).subs(point) * zj * zk

# Higher order remainder
F3 = F - F0 - F1 - F2

# Higher order remainder
F3 = F - F0 - F1 - F2

print('F0 =', F0)
print('F1 =', F1)
print('F2 =', F2)
print('F3 =', F3)
```

```python
# Create meshgrid
import numpy as np
N = 128
x1_range = np.linspace(0, 2*np.pi, N)
x2_range = np.linspace(0, 2*np.pi, N)
X1_grid, X2_grid = np.meshgrid(x1_range, x2_range)

# Lambdify each Taylor term (vectorized over x1,x2)
tar = lambdify((x1, x2, eps), F1[0, 0], 'numpy')

# Evaluate over entire meshgrid at once (4x1 matrix at each grid point)
tar_grid = tar(X1_grid, X2_grid, 1e-3)
```
